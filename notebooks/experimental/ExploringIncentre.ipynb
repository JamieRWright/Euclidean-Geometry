```python
# The coordinates of the incentre are the weighted average of the coordinates of the vertices, where
# the weights are the side lengths...
# T a triangle vertices A, B, C
# a = BC, b = CA, c = AB
# in vector language, the incentre is aA + bB + cC/(a+b+c)

R.<a, b, c, Ix, Iy, Ax, Ay, Bx, By, Cx, Cy> = PolynomialRing(QQ, 11, order="degrevlex(3),degrevlex(8)")

I = R.ideal([
    a^2 - (Bx-Cx)^2 - (By-Cy)^2,
    b^2 - (Cx-Ax)^2 - (Cy-Ay)^2,
    c^2 - (Ax-Bx)^2 - (Ay-By)^2,
    a*Ax + b*Bx + c*Cx - (a+b+c)*Ix,
    a*Ay + b*By + c*Cy - (a+b+c)*Iy
])
```

```python
B = I.groebner_basis()
```
